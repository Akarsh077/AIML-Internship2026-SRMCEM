```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Conv2D,MaxPooling2D,Flatten
```

```python
model=keras.Sequential()
model.add(Conv2D(32, (3, 3), padding='same', activation='relu', input_shape=(32,32,3)))
model.add(MaxPooling2D((2, 2)))

model.add(MaxPooling2D((2,2)))
model.add(Conv2D(64, (3,3), strides=(1,1), padding='same', activation='relu'))
model.add(MaxPooling2D((2,2)))
model.add(Flatten())
model.add(Dense(64, activation='relu'))

model.add(Dense(10, activation='softmax'))

model.summary()
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
predict_class = [53]   # Pizza class (COCO)

img = "./leo3.jpg"
pred = model(img, classes=predict_class, conf=0.4)

# Count detected pizzas
pizza_count = len(pred[0].boxes)

print("Number of pizzas detected:", pizza_count)

# Draw bounding boxes
pred_img = pred[0].plot()

import cv2

# Write the count on the image
cv2.putText(
    pred_img,
    f"Pizzas: {pizza_count}",
    (20, 40),
    cv2.FONT_HERSHEY_SIMPLEX,
    1,
    (0, 255, 0),
    2
)

cv2.imwrite("image2.jpg", pred_img)

# Display the image
cv2.imshow("Detected Pizzas", pred_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
